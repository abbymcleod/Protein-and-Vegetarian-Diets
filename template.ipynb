## Step 2: Data Cleaning and Exploratory Data Analysis

```python
interactions = pd.read_csv('/Users/abbymcleod/Desktop/DSC80/dsc80-2025-sp-1/projects/project04/interactions.csv')
recipes = pd.read_csv('/Users/abbymcleod/Desktop/DSC80/dsc80-2025-sp-1/projects/project04/RAW_recipes.csv')
```

```python
merged = pd.merge(recipes,interactions, how = 'left', left_on='id',right_on='recipe_id')
merged['rating'] = merged['rating'].replace(0,np.nan) #reasonable step because the 0 values probably represent missing or placeholder values, not actual data, so they will skew the average ratings
avg_rating = merged.groupby('id')['rating'].mean()
recipes['avg_rating'] = recipes['id'].map(avg_rating)
#recipes = recipes[recipes['minutes'] < 200]
recipes[['avg_rating','minutes']].sort_values(by='minutes', ascending = False)
```

```python
recipes['minute_bins'] = pd.cut(recipes['minutes'],bins = [0,15,30,60,120,240,300000],labels = ['0-15','16-20','31-60','61-120','121-240','241+'])
fig = px.box(recipes, x='minute_bins', y='avg_rating', title='Avg Rating by Recipe Prep Minutes',labels = {'minute_bin': 'Prep Time (minutes)','avg_rating':'Average Rating'})
fig.show()
```

```python
fig = px.histogram(recipes, x='avg_rating', nbins=20, title='Distribution of Average Ratings')
fig.show()
```
